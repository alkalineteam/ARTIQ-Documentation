
## ARTIQ installation
In a fresh Debian based installation (preferred Ubuntu 24.04), type in the following commands:

```bash
sudo apt install git
git clone https://github.com/alkalineteam/ARTIQ-alkaline.git cd ARTIQ-alkaline
sudo chmod +x setup.sh
./setup.sh  
```


Exit and re-open the shell.
- Git is where our ARTIQ repositories are located, and you are essentially making a copy of that repository in your local device
- The setup.sh is a shell script that will help you install curl, nix, enables flake and allows unfree packages for your nix installation.

<br>

To verify the nix installation, type in: 
```bash
nix --version
# you should see a version number
cat ~/.config/nix/nix.conf 
# you should see an output like this:
>> experimental-features = nix-command flakes 
>> extra-sandbox-paths = /opt
>> allow-unfree = true
```
<br>

Now that the environment is setup, it's time to switch branch to your desired crate. <br><br>
**Available branches are: <br> main ***(don't touch it)***, transportable, testbed, possible.**<br><br>
To switch branch, type in:
```bash
git checkout [branch_name]
```
<br>


Now it's time to get inside of the nix shell from where you can access all the ARTIQ commands

```bash
nix develop --impure
# takes a bit of time on the first run
```


Do a sanity check:
```bash
artiq_master --version
# you should see a version number
```
<br>


Now it's time to spin up the master:
```bash
artiq_master
```

<br>

Open another terminal and type in:
```bash
nix develop --impure
```
<br>
and once you're in the shell, get inside the dashboard

```bash
artiq_dashboard
```
<br>

You can store your experiments in the `repository` folder already available at the root of the directory. <br>
